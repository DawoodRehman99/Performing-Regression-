Using Label Encoding on Categorical Data.

```python
Label_encoder= LabelEncoder()
df["Brand"]=Label_encoder.fit_transform(df['Brand'])
df['Owner_Type']=Label_encoder.fit_transform(df['Owner_Type'])
df['Transmission']=Label_encoder.fit_transform(df['Transmission'])
df['Fuel_Type']=Label_encoder.fit_transform(df['Fuel_Type'])
```

```python
df.drop(['Model'],axis=1,inplace=True)
```

```python
df.head()
```

Features and Target..

```python
x=df.drop('Price', axis=1)
y=df['Price']
```

```python
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
def evaluate_models(x_train, y_train, x_test, y_test):
    #define Models in Dic form
    models={ 
    'Linear Regression': LinearRegression(),
    'Support Vector Regression':SVR(),
    'Decision Tree':DecisionTreeRegressor(), 
    'Gradient Boosting':GradientBoostingRegressor()
          }
   # Store results
    results = {} 
    
     # Evaluate each model
    for name, model in models.items():
        # Fit the model and make predictions
        model.fit(x_train, y_train)
        predictions = model.predict(x_test)
        
        # Calculate MSE and RÂ²
        mse = mean_squared_error(y_test, predictions)
        r2 = r2_score(y_test, predictions)

        results[name] = {
            'Test MSE': mse,
            'RÂ²': r2
        }

    return results

# Run the evaluation
results = evaluate_models(x_train, y_train, x_test, y_test)

# Print the results
for model_name, metrics in results.items():
    print(f"{model_name}: Test MSE = {metrics['Test MSE']:.2f}, RÂ² = {metrics['RÂ²']:.2f}")
```
